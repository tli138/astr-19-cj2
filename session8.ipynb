```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)


def sin(x):
    return np.sin(x)


def cos(x):
    return np.cos(x)
```

```python
y_sin = sin(x)
y_cos = cos(x)

fig, axs = plt.subplots(1, 2, figsize=(8, 4))


# Plot sin(x) vs. x in the left
axs[0].plot(x, y_sin)
axs[0].set_title('sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')


# Plot cos(x) vs. x in the right
axs[1].plot(x, y_cos)
axs[1].set_title('cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')

plt.savefig('figs/sin_and_cos_plot.pdf')
```
